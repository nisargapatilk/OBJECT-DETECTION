```python
#CascadeClassifier is used to load up the classifiers(pretrained model)
#detectMultiScale is used to identify the object detection in images
```

```python
img=cv2.imread("/content/p1.png")
faces=cv2.CascadeClassifier("/content/haarcascade_frontalface_default.xml")
f=faces.detectMultiScale(img,1.01,8)
for (x,y,w,h) in f:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
cv2_imshow(img)
```

```python
#CascadeClassifier is used to load up the classifiers(pretrained model)
#detectMultiScale is used to identify the object detection in images
```

```python
img=cv2.imread("/content/p2.png")
smiles=cv2.CascadeClassifier("/content/haarcascade_smile.xml.1")
s=smiles.detectMultiScale(img,1.9,20)
for (x,y,w,h) in s:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,0),4)
cv2_imshow(img)
```
